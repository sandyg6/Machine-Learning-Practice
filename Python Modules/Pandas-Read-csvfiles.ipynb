```python
data=('index,a,b,c\n'
      '4,apple,bat,5.7\n'
      '8,orange,cow,10')
```

```python
pd.read_csv(StringIO(data), index_col=0)  # making a column as index value
```

```python
pd.read_csv(StringIO(data)) 
```

```python
data=('index,a,b,c\n'
      '4,apple,5.7\n'
      '8,orange,cow')
```

```python
pd.read_csv(StringIO(data),index_col=0)  #default dtype=none so it takes as NaN
```

```python
pd.read_csv(StringIO(data),index_col=False)
```

```python
data=('index,a,b,c\n'
      '4,apple,bat,5.7\n'
      '8,orange,cow,10')
pd.read_csv(StringIO(data), usecols=['index','c','b'],index_col=0)
```

```python
pd.read_csv(StringIO(data), usecols=['index','c','b'],index_col=False)
```
