```python
import matplotlib.pyplot as plt

%matplotlib inline 
#while working in jupyter notebook we must use this
```

```python
import numpy as np
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
a=np.arange(40,50)
b=np.arange(50,60)
```

```python
#plotting using matplotlib

#plt scatter

plt.scatter(x,y,c='g')

#if we have'nt imported %matplotlib inline, we should've written a function plt.show()

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Graph 2D')
plt.savefig('test.png')  #saves the image of graph plotted
```

```python
y=x*x
```

```python
plt.plot(x,y,'g*',linestyle='dashed',linewidth=2,markersize=10)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("2D graph")
```

```python
## creating subplots

plt.subplot(2,3,1)  #(no. of rows, no. of cols, index)
plt.plot(x,y,'b*')
plt.subplot(2,3,2)
plt.plot(x,y,'r--')
plt.subplot(2,3,3)
plt.plot(x,y,'g*--')
plt.subplot(2,3,4)
plt.plot(x,y,'bo-')
plt.subplot(2,3,5)
plt.plot(x,y,'ro')
plt.subplot(2,3,6)
plt.plot(x,y,'g-*')
```

```python
# compute the xand y coordinates for points on a sine curve

x=np.arange(0,4*np.pi, 0.1)
y=np.sin(x)
plt.title("sine wave form")

#plot the points using matplotlib
plt.plot(x,y)
plt.show()
```

```python
#computing x and y cooordinates for points on sine and cosine curves
x=np.arange(0,5*np.pi,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(2,1,1)
plt.plot(x,y_sin,'g--')
plt.title('Sine')

plt.subplot(2,1,2)
plt.plot(x,y_cos,'b--')
plt.title('Cosine')

plt.show()
```

### Bar Chart

```python
#Bar chart

x=[2,8,10,5]
y=[11,16,9, 7]

x1=[3,9,11,7]
y1=[6,17,4,1]

plt.bar(x,y,align='center')
plt.bar(x1,y1, color='g')
plt.ylabel('Y-axis')
plt.xlabel('X-axis')

```

## Box Plot using Matplotlib

```python
data=[np.random.normal(0,std,100) for std in range(1,4)]

#rectangular box plot
plt.boxplot(data, vert= True, patch_artist=True);
```

```python
data=[np.random.normal(0,std,100) for std in range(1,4)]

#rectangular box plot
plt.boxplot(data, vert= False, patch_artist=False);
```

## Pie chart

```python
#Data to plot
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215,120,180,240]
colors = ['magenta', 'seagreen', 'lightgrey', 'whitesmoke']
explode=(0,0,0,0.1)  #explode 1st slide
#plt 
plt.pie(sizes, explode=explode,labels=labels,colors=colors, 
        autopct='%1.3f%%',shadow=True)

plt.axis('equal')
```
