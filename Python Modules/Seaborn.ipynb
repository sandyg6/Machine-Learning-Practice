# Seaborn

### Distribution plots
- displot
- joinplot
- pairplot

```python
import seaborn as sns
```

```python
df=sns.load_dataset("tips")
```

```python
df.head()

# tip is a dependent feature
# independent feature - total_bill, sex, smoker, day,time, size
```

```python
df.dtypes
```

# Correlation with Heatmap
It helps to find inter relation between all the features

```python
df.corr(numeric_only=True)
```

```python
sns.heatmap(df.corr(numeric_only=True))
```

## Joint plot

### Univarient analysis

```python
sns.jointplot(x='tip',y='total_bill', data=df, kind='reg',dropna=True)
```

```python
sns.jointplot(x='tip',y='total_bill', data=df, kind='hex')
```

## Pair plot
Also known as scatter plot

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='day')
```

```python
sns.pairplot(df,hue='sex')
```

```python
sns.pairplot(df,hue='smoker')
```

```python
df['smoker'].value_counts()
```

## Dist plot

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['total_bill'])
```

```python
sns.distplot(df['tip'],kde=False,bins=10) 
#kde- kernal density estimation
```

## Categorical plot

```python
#count plot

# sns.countplot('sex',data=df)
sns.countplot(x='sex', y=None,data=df)
#shows counts of observations in each categorical bin using bars
```

```python
sns.countplot(x='smoker', y=None,data=df)
```

```python
sns.countplot(x='day', y=None,data=df)
```

```python
sns.countplot(x=None,y='sex',data=df)
```

```python
sns.countplot(x=None,y='smoker',data=df)
```

## Bar plot

```python
sns.barplot(x='total_bill',y='sex',data=df)
```

```python
sns.barplot(x='smoker',y='total_bill',data=df)
```

```python
sns.barplot(x='sex',y='total_bill',data=df)
```

## Box plot
Abox and whisker plot (sometimes called a boxplot) is a graph that presents information from a five-number summary.

```python
sns.boxplot(x='smoker',y='total_bill',data=df)
```

```python
sns.boxplot(x='sex',y='total_bill',data=df)
```

```python
sns.boxplot(x='day',y='total_bill',data=df, palette='rainbow')
```

```python
sns.boxplot(data=df,orient='h')
```

```python
sns.boxplot(data=df,orient='v')
```

```python
sns.boxplot(x='total_bill',y='day',hue='smoker',data=df)
```

```python
sns.boxplot(x='total_bill',y='day',hue='sex',data=df)
```

## Violin plot

```python
sns.violinplot(x='total_bill',y='day',data=df,palette='rainbow')
```
