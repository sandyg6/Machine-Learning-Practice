```python
import pandas as pd
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import accuracy_score
from xgboost import XGBClassifier
from sklearn.impute import SimpleImputer

# 1. Load the dataset
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')

# 2. Handle missing values
# Use SimpleImputer for missing values handling
imputer = SimpleImputer(strategy='median')
train['Age'] = imputer.fit_transform(train[['Age']])
test['Age'] = imputer.transform(test[['Age']])

# Impute missing Embarked and Fare with mode and median respectively
train['Embarked'] = train['Embarked'].fillna(train['Embarked'].mode()[0])
test['Fare'] = test['Fare'].fillna(test['Fare'].median())

# Create a binary feature for missing cabins (Has_Cabin)
train['Has_Cabin'] = train['Cabin'].notna().astype(int)
test['Has_Cabin'] = test['Cabin'].notna().astype(int)

# 3. Feature Engineering: Extract Titles and FamilySize
train['Title'] = train['Name'].apply(lambda x: x.split(',')[1].split('.')[0].strip())
test['Title'] = test['Name'].apply(lambda x: x.split(',')[1].split('.')[0].strip())

# Group rare titles into 'Other'
rare_titles = ['Dr', 'Rev', 'Major', 'Capt', 'Col', 'Sir', 'Don', 'Dona', 'Lady', 'Countess', 'Jonkheer', 'Baroness', 'Mme', 'Ms', 'Mlle', 'Mr']
train['Title'] = train['Title'].replace(rare_titles, 'Other')
test['Title'] = test['Title'].replace(rare_titles, 'Other')

# Create a FamilySize feature
train['FamilySize'] = train['SibSp'] + train['Parch']
test['FamilySize'] = test['SibSp'] + test['Parch']

# 4. Encoding Categorical Features
# Label encode 'Sex' (Male = 1, Female = 0)
le = LabelEncoder()
train['Sex'] = le.fit_transform(train['Sex'])
test['Sex'] = le.transform(test['Sex'])

# One-hot encode 'Embarked' and 'Title'
train = pd.get_dummies(train, columns=['Embarked', 'Title'], drop_first=True)
test = pd.get_dummies(test, columns=['Embarked', 'Title'], drop_first=True)

# 5. Feature Scaling: Standardize numerical features
scaler = StandardScaler()
train[['Age', 'Fare', 'FamilySize']] = scaler.fit_transform(train[['Age', 'Fare', 'FamilySize']])
test[['Age', 'Fare', 'FamilySize']] = scaler.transform(test[['Age', 'Fare', 'FamilySize']])

# 6. Align columns between training and test data
# Drop unnecessary columns like 'PassengerId', 'Name', 'Cabin', and 'Ticket'
X_train = train.drop(['Survived', 'PassengerId', 'Name', 'Cabin', 'Ticket'], axis=1)
X_test = test.drop(['PassengerId', 'Name', 'Cabin', 'Ticket'], axis=1)

# Align the test set to have the same columns as the training set
# Add any missing columns to the test set (if any) and remove extra columns in test
X_test = X_test.reindex(columns=X_train.columns, fill_value=0)

# Define the target variable
y_train = train['Survived']
y_test = test['Survived']

# 7. Hyperparameter Tuning with RandomizedSearchCV for RandomForestClassifier
rf = RandomForestClassifier(random_state=42)

# Randomized search hyperparameters
param_dist = {
    'n_estimators': [100, 200, 300, 500],
    'max_depth': [3, 5, 7, 10, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2', None],
    'bootstrap': [True, False]
}

# RandomizedSearchCV for hyperparameter optimization
rf_random_search = RandomizedSearchCV(rf, param_distributions=param_dist, n_iter=100, cv=5, verbose=1, n_jobs=-1, scoring='accuracy', random_state=42)
rf_random_search.fit(X_train, y_train)

# Best Random Forest Model
best_rf = rf_random_search.best_estimator_

# 8. Train an XGBoost model for comparison
xgb = XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='logloss')
xgb.fit(X_train, y_train)

# 9. Make predictions using both models
rf_pred = best_rf.predict(X_test)
xgb_pred = xgb.predict(X_test)

# 10. Evaluate the models on the test set
rf_accuracy = accuracy_score(y_train, best_rf.predict(X_train))
xgb_accuracy = accuracy_score(y_train, xgb.predict(X_train))

# 11. Select the best model (based on accuracy on training data)
best_model = best_rf if rf_accuracy > xgb_accuracy else xgb
best_pred = rf_pred if rf_accuracy > xgb_accuracy else xgb_pred

# 12. Create the final submission file
submission = pd.DataFrame({'PassengerId': test['PassengerId'], 'Survived': best_pred})
submission.to_csv('titanic_submission_final.csv', index=False)

# Output training accuracies for comparison
print(f"Random Forest Training Accuracy: {rf_accuracy:.4f}")
print(f"XGBoost Training Accuracy: {xgb_accuracy:.4f}")
```
